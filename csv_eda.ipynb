```python
from pathlib import Path
import pandas as pd
```

```python
csv_dir = Path.cwd() / 'eduscores' / 'data' / 'csv'

def get_csv(year, entities=True):
    """2015-2018"""
    
    year = str(year)
    for path in csv_dir.iterdir():
        if year not in path.name:
            continue
            
        if entities:
            if 'entities' in path.name:
                return path
        else:
            if 'all_csv' in path.name:
                return path
    raise Exception(f'year {year} not in files')
```

```python
sbac_csv = get_csv(2018, entities=False)

sbac_df = pd.read_csv(sbac_csv, encoding='latin1')
sbac_df.columns = [col.lower().replace(' ', '_') for col in sbac_df.columns]
sbac_df = (sbac_df.drop(columns=['filler'] + [col for col in sbac_df.columns 
                                              if any(s in col for s in ['area', 'total', 'percent'])])
                  .set_index(['school_code', 'subgroup_id', 'grade', 'test_id', 'test_year'])
                  .sort_index())

sbac_df[sbac_df.index.get_level_values(0) != 0]
```

```python
entities_csv = get_csv(2018, entities=True)

entities_df = pd.read_csv(entities_csv, encoding='latin1')
entities_df.columns = [col.lower().replace(' ', '_') for col in entities_df.columns]
entities_df = (entities_df.drop(columns=['filler'])
                          .set_index(['school_code', 'test_year'])
                          .sort_index())

entities_df[entities_df.index.get_level_values(0) != 0]
```
